```python
#Implement a function longestWord() that takes a list of words and returns the longest one.
def longestWord(str_list):
    word_max_length = 0
    for word in str_list.split():
        if len(word) > word_max_length:
            word_max_length = len(word)
    print ("The longest word is", word, "with length", word_max_length)


str_list = input("Enter the list of words: ")
longestWord(str_list)
```
